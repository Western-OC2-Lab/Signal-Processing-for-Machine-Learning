```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import loadtxt
import scipy
from scipy import signal
plt.rc('font',family='Times New Roman')
import warnings 
warnings.filterwarnings('ignore')
```

```python
# load vibration signal:
random_vib_signal = loadtxt('vib.csv', delimiter=',') 
num_samples_vib = len(random_vib_signal) # number of datapoints in the signal.
fs_vib = 64000 # sampling rate
t_vib = np.arange(0.0, num_samples_vib, 1.0)/fs_vib # time vector.
```

#The vibration signal extracted from the PU dataset: https://mb.uni-paderborn.de/en/kat/main-research/datacenter/bearing-datacenter/data-sets-and-download

```python
seg1= 300
seg2 = 1000

signal = random_vib_signal[seg1:seg2]
time = t_vib[seg1:seg2]

# Plot the Power Spectral Density
plt.figure(figsize=(20, 15))
plt.plot(time, signal)
plt.title('$x(t)$', fontsize = 40)
plt.xlabel('Time ($t$)', fontsize = 40)
plt.ylabel('Amplitude', fontsize = 40)
plt.tick_params(axis='both', which='major', labelsize=40)
plt.tick_params(axis='both', which='minor', labelsize= 40)
plt.xticks([])  
plt.show()
```

```python

autocorr = np.correlate(signal, signal, mode='same')
lags = np.arange(-len(signal) // 2 + 1, len(signal) // 2 + 1)

# Plotting
plt.figure(figsize=(20, 15))

# Plot the original signal
plt.subplot(3, 1, 1)
plt.plot(time, signal)
plt.title('$x(t)$',  fontsize = 40)
plt.xlabel('Time ($t$)',  fontsize = 40)
plt.ylabel('Amplitude',  fontsize = 40)
plt.tick_params(axis='both', which='major', labelsize=40)
plt.tick_params(axis='both', which='minor', labelsize= 40)

# Plot the autocorrelation of the signal
plt.subplot(3, 1, 2)
plt.plot(lags[len(signal)//2:], autocorr[len(signal)//2:])
plt.title('Autocorrelation of $x(t)$',  fontsize = 40)
plt.xlabel('Lag',  fontsize = 40)
plt.ylabel('Autocorrelation',  fontsize = 40)
plt.tick_params(axis='both', which='major', labelsize=40)
plt.tick_params(axis='both', which='minor', labelsize= 40)

plt.subplot(3, 1, 3)
plt.plot(lags, autocorr)
plt.title('Autocorrelation of $x(t)$',  fontsize = 40)
plt.xlabel('Lag',  fontsize = 40)
plt.ylabel('Autocorrelation',  fontsize = 40)
plt.tick_params(axis='both', which='major', labelsize=40)
plt.tick_params(axis='both', which='minor', labelsize= 40)

plt.grid(True)
plt.tight_layout()
plt.show()
```
